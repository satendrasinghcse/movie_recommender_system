```python
df = data1.merge(data2,on="title")
```

```python
df = df[['movie_id','title','overview','genres','keywords','cast','production_companies','crew']]
```

```python
df.head()
```

```python
pd.set_option('max_colwidth',None)
#pd.reset_option(“max_columns”)
df["genres"]
pd.reset_option("max_colwidth",None)
```

```python
df["genres"][0]
```

```python
import json

def extract(str):
    text_list = json.loads(str)
    text = [genre['name'] for genre in text_list]
    return text

```

```python
df["keywords"] = df["keywords"].apply(extract)
df["genres"] = df["genres"].apply(extract)
```

```python
def extract3(str):
    text_list = json.loads(str)
    text = [genre['name'] for genre in text_list]
    return text[:3]
```

```python
df["cast"] = df["cast"].apply(extract3)
```

```python
df["production_companies"] = df["production_companies"].apply(extract3)
```

```python
df["crew"][0]
```

```python

def find_directors(data):
    name = []
    data = json.loads(data)
    for i in data:
        if i['job']=='Director':
            name.append(i["name"])
    return name
```

```python
df["crew"] = df["crew"].apply(find_directors)
```

```python
df.head()
```

```python
df["title"].apply(lambda x:x.split())
```

```python
df["overview"] = df["overview"].apply(lambda x:str(x).split())
```

```python
df["tags"] = df["title"].apply(lambda x:x.split())+df["cast"]+df["crew"]+df["genres"]+df["keywords"]+df["overview"]+df["production_companies"]
```

```python
df.head()
```

```python
df.columns
```

```python
new_df = df.drop(columns=['overview', 'genres', 'keywords', 'cast',
       'production_companies', 'crew'])
```

```python
new_df.head()
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
from sklearn.feature_extraction.text import TfidfVectorizer
cv1 = TfidfVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
vector1 = cv1.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity1 = cosine_similarity(vector1)
```

```python
similarity
```

```python
new_df[new_df['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new_df[new_df['title'].str.lower() == str(movie).lower()].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)


def recommend1(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity1[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
print(recommend('Gandhi'))
print("********************************************")
print(recommend1("Gandhi"))
```
